```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xi = [-0.9, 0., 0.9, 1.8, 2.7, 3.6]
yi = [-1.2689, 0., 1.2689, 2.6541, 4.4856, 9.9138]

x_smooth = np.linspace(min(xi) - 0.5, max(xi) + 0.5, 100)
y_poly1 = -0.190129 + 2.246206 * x_smooth
y_poly2 = -0.464496 + 0.874367 * x_smooth + 0.508089 * x_smooth**2


plt.figure(figsize=(10, 6))
plt.scatter(xi, yi, color='black', s=50, zorder=5, label='Source points (xi, yi)')

plt.plot(x_smooth, y_poly1, 'r-', linewidth=2, label='y = -0.190129 + 2.246206*x')
plt.plot(x_smooth, y_poly2, 'b-', linewidth=2, label='y = -0.464496 + 0.874367*x + 0.508089*x^2')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Data approximation by polynoms')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()

plt.show()
```
